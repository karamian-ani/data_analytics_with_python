```python
import pandas as pd
from scipy import stats
import plotly.express as px
```

```python
df = pd.read_csv('../data/ab_testing/fast_food.csv')
df.head()
```

```python
df.describe()
```

```python
df['MarketID'].value_counts()
```

```python
df['MarketSize'].value_counts()
```

```python
fig = px.box(
    df,
    x = "Promotion",
    y = "SalesInThousands",
    color = "Promotion",
    title = "Sales Distribution by Promotion"
)

fig
```

```python
weekly_sales = (
    df.groupby(["week","Promotion"])['SalesInThousands'].mean().reset_index()
)
```

```python
weekly_sales.head()
```

```python
px.line(
    data_frame = weekly_sales,
    x = "week",
    y = "SalesInThousands",
    color = "Promotion"

)
```

```python
px.box(
    data_frame = df,
    x = 'MarketSize',
    y = 'SalesInThousands',
    color = "Promotion",
    title = "Sales By MarketSize per Promotion"
)
```

```python
fig = px.scatter(
    df,
    x="AgeOfStore",
    y="SalesInThousands",
    color="Promotion",
    trendline="ols",
    title="Sales vs Store Age"
)
fig.show()
```
